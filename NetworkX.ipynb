```python
# Declaring two graphs
Q1 = nx.complete_graph(5)
Q2 = nx.complete_graph(5)

#renaming nodes of graph Q2
Q2 = nx.relabel_nodes(Q2, {0:"A", 1:"B", 2:"C", 3:"D", 4:"E"})

# Setting up connection between two graphs
# Node '4' will connect to node "X" and node "X" will connect to node "A"
Q_connector = nx.from_edgelist([(4, "X"), ("X", "A")])

Q = nx.compose_all([Q1, Q2, Q_connector])

nx.draw_spring(Q,with_labels = True)
plt.show()
```

```python
print(nx.degree_centrality(Q))
```

```python
print(nx.betweenness_centrality(Q))
```

```python
print(nx.density(Q))
```

```python
print(nx.diameter(Q))
```

```python
print(list(nx.eulerian_path(Q)))
```

```python
print(list(nx.find_cliques(Q)))
```

```python
print(list(nx.bridges(Q)))
```

```python
# Declaring two graphs
s1 = nx.complete_graph(5)
s2 = nx.complete_graph(5)

#renaming nodes of graph Q2
s2 = nx.relabel_nodes(s2, {0:"A", 1:"B", 2:"C", 3:"D", 4:"E"})

# Setting up connection between two graphs
# Node '4' will connect to node "X" and node "X" will connect to node "A"
s_connector = nx.from_edgelist([(4, "X"), ("X", "A"),
                                (4, "F"), ("F", "G"),
                                ("G", "H"), ("H", "A")])

s = nx.compose_all([s1, s2, s_connector])

nx.draw_spring(s,with_labels = True)
plt.show()
```

```python
print(list(nx.bridges(s)))
```

### **Multiple Edge Attributes**
> * Create adjacency matrices for multiple edge attributes with structured dtypes:

```python
F = nx.Graph()
F.add_edge(0, 1, weight = 10)
F.add_edge(1, 2, cost = 5)
F.add_edge(2, 3, weight = 3, cost = -4.0)
dtype = np.dtype([("weight", int), ("cost", float)])
A = nx.to_numpy_array(F, dtype = dtype, weight = None)
print("Adjacency matrix of weight ------")
print(A["weight"],"\n")
print("Adjacency marix of cost ------")
print(A["cost"], "\n")
nx.draw(F, with_labels = True)
print(f"Nodes in graph {F.nodes}\n")
print(f"Edges in graph {F.edges}\n")
```
